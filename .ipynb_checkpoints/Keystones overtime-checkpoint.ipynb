```python
X = []
Y = []
Q = [] #time test
J = [] #rank
```

```python

with open('data/test.json') as f:
    data = json.load(f)
    
for i in range(100):
    X.append(data['leading_groups'][i]['keystone_level'])
    ts = int(data['leading_groups'][i]['duration'])
    Y.append(ts)
    Q.append(data['leading_groups'][i]['completed_timestamp'])
    J.append(data['leading_groups'][i]['ranking'])

```

```python
plt.plot(X, Y)
plt.show()
```
